```python
import math
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 50)
y = np.cos(x)
 
print("in_array : ", x)
print("\nout_array : ", y)
 
# red for numpy.cos()
plt.plot(x, y, color = 'red', marker = "o")
plt.title("numpy.cos()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
def fun(x):
  return x**2;

x = np.linspace(0, 2*np.pi, 50)
y =[fun(x) for x in x]
print(y)
plt.plot(x, y, color = 'green', marker = "o")
plt.title("x^2")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()


```

4. Use the given Taylor series as an exponential function. (Refer to the image for e^x)
a. Compute the error of Taylor series when using 10 and 15 terms for x = 2.
i. Compute e^x for 10 terms or 15 terms using the series
ii. Compute e^x for x=2 directly
iii. Compare the result of i and ii
b. Find the appropriate number of terms when the error can be in a range of 0.001 for
x = 3.

```python
#	x**0/math.factorial(0)
import math
import numpy as np

sum=0.0
x=2
for i in range(10):
  	sum += float(x**i/math.factorial(i))
print(sum)   
calculated_value=sum
```

```python
actual_value=math.exp(2)
```

```python
print(f"Error rate is :{actual_value-calculated_value} ")
```

b. Find the appropriate number of terms when the error can be in a range of 0.001 for
x = 3.

```python
sum=0
x=3
i=0
actual_value=math.exp(x)
while (actual_value - sum) >= 0.001:
  sum += float(x**i/math.factorial(i))
  i+=1
print(i)
print(actual_value-sum)  
```
